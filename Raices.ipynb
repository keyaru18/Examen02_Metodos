```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize

def f(x):
    return (x**5 + 5.5*x**4 + 4.5*x**3 - 9.5*x**2 - 15.5*x - 6)
# Lista de intervalos donde hay cambio de signo 
intervals = [(-10, -4), (-4, -2), (-2, 0), (0.5, 2)]
# Lista para almacenar las raíces
roots = []
for a, b in intervals:
    if f(a) * f(b) < 0:  # Verificar cambio de signo
        root = optimize.bisect(f, a, b, xtol=1e-12, rtol=1e-12, maxiter=100)
        roots.append(round(root, 10))  # Redondear para evitar decimales infinitos
    else:
        roots.append(np.nan)
# Rellenar con NaN hasta tener 6 valores
while len(roots) < 6:
    roots.append(np.nan)    

# Imprimir raíces
for i, r in enumerate(roots, 1):
    print(f"x{i} = {r}")    



```
